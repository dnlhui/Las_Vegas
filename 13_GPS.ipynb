```python
import pandas as pd
import numpy as np
```

### Load CSV

```python
data_df = pd.read_csv("../02_Data/07_Predictions/z_prediction.csv",index_col=0)
data_df.head()
```

### Split Code to Row + Column

```python
def get_grid(row):
    row = str(row)
    column_coord = row[-3:]
    row_coord = row[:-3]
    return row_coord, column_coord
```

```python
get_grid("4052")
```

```python
data_df["row"] = data_df["code"].apply(get_grid)                 # Get Tuple of Row, Column
data_df["column"] = data_df["row"].apply(lambda x: int(x[1]))    # Grab Column from Tuple
data_df["row"] = data_df["row"].apply(lambda x: int(x[0]))       # Grab Row from Tuple
data_df.head()
```

### Find Rows/Columns for GPS Cross-Check
Look for extreme min and max for rows and columns to correspond the GPS coordinates back to the images

```python
data_df[data_df["row"]==28].head()
```

### GPS Outputs
Refer to Excel Spreadsheet "GPS" for more detail

Row = Latitude

Column = Longitude

```python
lat_per_image = -0.00181471370967745
long_per_image = 0.00224789423076922
```

```python
lat_origin = 36.340694733871
long_origin = -115.385946384615
```

```python
data_df["lat"] = data_df["row"].apply(lambda x: x*lat_per_image + lat_origin + lat_per_image/2)  
data_df["long"] = data_df["column"].apply(lambda x: x*long_per_image + long_origin + long_per_image/2)

# last /2 is for center point of lat & long
```

```python
data_df.head()
```

### Export to CSV

```python
export_df = data_df[["code","zoning","predicted_zoning","lat","long"]]
export_df.head()
```

```python
export_df.to_csv("../09_Tableau/gps.csv")
```
